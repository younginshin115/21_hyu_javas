# 토큰화

```python
# 불용어 정의 (한국어 조사, 접속사 등) 
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']
```

```python
import json
with open('KoNLPy(Mecab)_stopwords(18)_Total(movie&shopping&game&target)_tokened_vocab_v0.1.json', encoding='UTF8') as json_file:
    X_total_10X = json.load(json_file)
```

# 정수 인코딩

```python
vocab_size = 9237
tokenizer_10X = Tokenizer(num_words = vocab_size +2, oov_token ='OOV')
```

```python
import json
with open('KoNLPy(Mecab)_stopwords(18)_Total(movie&shopping&game&target)_indexed_vocab_v0.1.json', encoding='UTF8') as json_file:
#   vocab = json.load(json_file)
    tokenizer_10X.word_index = json.load(json_file)
```

```python
# tokenizer_10X = Tokenizer(num_words = vocab_size +2, oov_token ='OOV')
# tokenizer_10X.fit_on_texts(X_total_10X)
X_total_10X = tokenizer_10X.texts_to_sequences(X_total_10X)
y_total_10X = np.array(total_data_10X['label'])
```

```python
X_total_10X [:3]
```

# 빈 샘플(empty samples)제거

```python
drop_total_10X = [index for index, sentence in enumerate(X_total_10X) if len(sentence) < 1]
X_total_10X = np.delete(X_total_10X, drop_total_10X, axis=0)
y_total_10X = np.delete(y_total_10X, drop_total_10X, axis=0)
```

# 패딩

```python
max_len = 45
X_total_10X = pad_sequences(X_total_10X, maxlen = max_len)
```

#  모델 훈련

```python
# 필요한 프레임워크 및 라이브러리 임포트
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Embedding, Dropout, Conv1D, GlobalMaxPooling1D, Dense, Input, Flatten, Concatenate
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.models import load_model
```

```python
# 필요한 매트릭 선언, AUC-ROC, F1-Score, AUC-PR 에 필요한 지표 모두 체크
import tensorflow as tf
ROCauc = tf.keras.metrics.AUC(curve="ROC") # auc
PRauc = tf.keras.metrics.AUC(curve="PR") # auc_1
RECALL = tf.keras.metrics.Recall()
PRECISION = tf.keras.metrics.Precision()
TP = tf.keras.metrics.TruePositives()
TN = tf.keras.metrics.TrueNegatives()
FP = tf.keras.metrics.FalsePositives()
FN = tf.keras.metrics.FalseNegatives()
```

```python
embedding_dim = 128
dropout_prob = (0.1, 0.1)
num_filters1 = 64
num_filters2 = 32
num_filters3 = 16
num_filters4 = 256
```

```python
model_input = Input(shape = (max_len,))
z = Embedding(vocab_size+2, embedding_dim, input_length = max_len, name="embedding")(model_input)
z = Dropout(dropout_prob[0])(z)
```

```python
conv_blocks1 = []
for sz in [3, 4, 5]:
    conv1 = Conv1D(filters = num_filters1,
                         kernel_size = sz,
                         padding = "valid",
                         activation = "relu",
                         strides = 1)(z)
    conv1 = GlobalMaxPooling1D()(conv1)
    conv1 = Flatten()(conv1)
    conv_blocks1.append(conv1)
```

```python
conv_blocks2 = []
for sz in [3, 4, 5]:
    conv2 = Conv1D(filters = num_filters2,
                         kernel_size = sz,
                         padding = "valid",
                         activation = "relu",
                         strides = 1)(z)
    conv2 = GlobalMaxPooling1D()(conv2)
    conv2 = Flatten()(conv2)
    conv_blocks2.append(conv2)
```

```python
conv_blocks3 = []
for sz in [3, 4, 5]:
    conv3 = Conv1D(filters = num_filters3,
                         kernel_size = sz,
                         padding = "valid",
                         activation = "relu",
                         strides = 1)(z)
    conv3 = GlobalMaxPooling1D()(conv3)
    conv3 = Flatten()(conv3)
    conv_blocks3.append(conv3)
```

```python
conv_blocks4 = []
for sz in [3, 4, 5]:
    conv4 = Conv1D(filters = num_filters4,
                         kernel_size = sz,
                         padding = "valid",
                         activation = "relu",
                         strides = 1)(z)
    conv4 = GlobalMaxPooling1D()(conv4)
    conv4 = Flatten()(conv4)
    conv_blocks4.append(conv4)
```

```python
z1 = Concatenate()(conv_blocks1) if len(conv_blocks1) > 1 else conv_blocks1[0]
z1 = Dropout(dropout_prob[1])(z1)
z1 = Dense(128, activation="relu")(z1)
model_output1 = Dense(1, activation="sigmoid")(z1)
model_z1 = Model(model_input, model_output1)
```

```python
z2 = Concatenate()(conv_blocks2) if len(conv_blocks2) > 1 else conv_blocks2[0]
z2 = Dropout(dropout_prob[1])(z2)
z2 = Dense(128, activation="relu")(z2)
model_output2 = Dense(1, activation="sigmoid")(z2)
model_z2 = Model(model_input, model_output2)
```

```python
z3 = Concatenate()(conv_blocks3) if len(conv_blocks3) > 1 else conv_blocks3[0]
z3 = Dropout(dropout_prob[1])(z3)
z3 = Dense(128, activation="relu")(z3)
model_output3 = Dense(1, activation="sigmoid")(z3)
model_z3 = Model(model_input, model_output3)
```

```python
z4 = Concatenate()(conv_blocks4) if len(conv_blocks4) > 1 else conv_blocks4[0]
z4 = Dropout(dropout_prob[1])(z4)
z4 = Dense(128, activation="relu")(z4)
model_output4 = Dense(1, activation="sigmoid")(z4)
model_z4 = Model(model_input, model_output4)
```

```python
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=4)
```

```python
model_z1.compile(loss="binary_crossentropy", optimizer="adam", metrics=['acc', ROCauc, PRauc, TP, TN, FP, FN])
mc_z1 = ModelCheckpoint('12_CNN_1D_KoNLPy(Mecab)_stopwords(18)_Total(movie&shopping&game&target10X)_v0.1.1.h5', monitor='val_acc', mode='max', verbose=1, save_best_only=True)
history = model_z1.fit(X_total_10X, y_total_10X, epochs=30, callbacks=[es, mc_z1], batch_size=100, validation_split=0.2)
```

```python
model_z2.compile(loss="binary_crossentropy", optimizer="adam", metrics=['acc', ROCauc, PRauc, TP, TN, FP, FN])
mc_z2 = ModelCheckpoint('12_CNN_1D_KoNLPy(Mecab)_stopwords(18)_Total(movie&shopping&game&target10X)_v0.1.2.h5', monitor='val_acc', mode='max', verbose=1, save_best_only=True)
history = model_z2.fit(X_total_10X, y_total_10X, epochs=30, callbacks=[es, mc_z2], batch_size=100, validation_split=0.2)
```

```python
model_z3.compile(loss="binary_crossentropy", optimizer="adam", metrics=['acc', ROCauc, PRauc, TP, TN, FP, FN])
mc_z3 = ModelCheckpoint('12_CNN_1D_KoNLPy(Mecab)_stopwords(18)_Total(movie&shopping&game&target10X)_v0.1.3.h5', monitor='val_acc', mode='max', verbose=1, save_best_only=True)
history = model_z3.fit(X_total_10X, y_total_10X, epochs=30, callbacks=[es, mc_z3], batch_size=100, validation_split=0.2)
```

```python
model_z4.compile(loss="binary_crossentropy", optimizer="adam", metrics=['acc', ROCauc, PRauc, TP, TN, FP, FN])
mc_z4 = ModelCheckpoint('12_CNN_1D_KoNLPy(Mecab)_stopwords(18)_Total(movie&shopping&game&target10X)_v0.1.4.h5', monitor='val_acc', mode='max', verbose=1, save_best_only=True)
history = model_z4.fit(X_total_10X, y_total_10X, epochs=30, callbacks=[es, mc_z4], batch_size=100, validation_split=0.2)
```

# 모델 성능평가(타겟데이터 기반)

```python
mecab = Mecab()
```

```python
# 평가용 토탈 데이터셋2(화장품+앱)

# 데이터 로드
total_data2 = pd.read_table('clean_total_data(app&cosmetic).txt')

# 토크나이징
X_test_total2 = []
for sentence in total_data2['document']:
    temp_X2 = mecab.morphs(sentence)
    temp_X2 = [word for word in temp_X2 if not word in stopwords]
    X_test_total2.append(temp_X2)

# 워드 벡터화
tokenizer = Tokenizer(vocab_size)
tokenizer.fit_on_texts(X_test_total2)
X_test_total2 = tokenizer.texts_to_sequences(X_test_total2)

# 패딩
X_test_total2 = pad_sequences(X_test_total2, maxlen = max_len)

# 라벨데이터 포함
y_test_total2 = np.array(total_data2['label'])
```

```python
import json
json = json.dumps(X_test_total2, ensure_ascii=False)
padded_data = open("KoNLPy(Mecab)_target(app&cosmetic)_padded_vocab_v0.1.json", "w", encoding="utf-8")
padded_data.write(json).close()
```
