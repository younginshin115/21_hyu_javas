```python
# 필요 프레임워크 및 라이브러리 준비
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

# 데이터 로드

```python
# 판다스로 각 파일데이터를  train/test 변수에 바인딩
shopping_data = pd.read_table('clean_naver_shopping_review_total_data.txt')
movie_data = pd.read_table('clean_naver_movie_review_total_data.txt')
game_data = pd.read_table('clean_steam_game_review_total_data.txt')
```

```python
# 데이터 체크
print('쇼핑 데이터 개수 :',len(shopping_data))
print('영화 데이터 개수 :',len(movie_data))
print('게임 데이터 개수 :',len(game_data))
```

```python
# 상위 5개 출력
shopping_data[:5]
# movie_data[:5]
# game_data[:5]
```

```python
# 데이터 프레임 합치기
total_data = pd.concat([shopping_data, movie_data, game_data])
print(len(total_data))
```

```python
total_data[:]
```

```python
total_data.drop(['id', 'ratings'], axis='columns', inplace=True)
total_data[:5]
```

```python
 # shuffling
total_data = total_data.sample(frac=1).reset_index(drop=True) 
total_data[:5]
```

# 데이터 클렌징

```python
# 중복 데이터 체크
total_data['document'].nunique(), total_data['label'].nunique()
```

```python
# 중복 데이터 제거
total_data.drop_duplicates(subset=['document'], inplace=True) # document 열에서 중복인 내용이 있다면 중복 제거
```

```python
# 중복 샘플 제거 확인
print('총 샘플의 수 :',len(total_data))
```

```python
# NULL값 유무 체크
print(total_data.isnull().values.any())
```

```python
# 훈련 데이터에서의 레이블 값의 분포 확인
total_data['label'].value_counts().plot(kind = 'bar')
```

```python
# 균일해보이지만 정확하게 몇개인지 확인해보자
print(total_data.groupby('label').size().reset_index(name = 'count'))
```

# 정제&정규화한 데이터 파일로 저장

```python
total_data.to_csv('total_data.txt', index=False, sep='\t')
```
