```python
# 필요 프레임워크 및 라이브러리 준비
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import sklearn
```

# 데이터 로드

```python
# 판다스로 각 파일데이터를  train/test 변수에 바인딩
target_data = pd.read_table('/content/drive/MyDrive/Colab Notebooks/NLP_Classification/sentiment/Cleaned Dataset/clean_total_data(app&cosmetic).txt')
train_data = pd.read_table('/content/drive/MyDrive/Colab Notebooks/NLP_Classification/sentiment/Cleaned Dataset/clean_total_data(movie&shopping&game).txt')
```

```python
# 데이터 체크
print('타겟데이터 수 :',len(target_data))
print('훈련데이터 수 :',len(train_data))
```

# 데이터 분리 및 체크

```python
# 타겟 데이터 셔플& 분리,  X_test를 훈련데이터에 섞을 예정
from sklearn.model_selection import train_test_split
X_train, X_test_10 = train_test_split(target_data, test_size=0.1, random_state=1004)
X_train, X_test_25 = train_test_split(target_data, test_size=0.25, random_state=1004)
X_train, X_test_50 = train_test_split(target_data, test_size=0.5, random_state=1004)
```

```python
print(len(X_test_10))
print(X_test_10.groupby('label').size().reset_index(name = 'count'))
print(X_test_10.head())
X_test_10['label'].value_counts().plot(kind = 'bar')
```

```python
print(len(X_test_25))
print(X_test_25.groupby('label').size().reset_index(name = 'count'))
print(X_test_25.head())

X_test_25['label'].value_counts().plot(kind = 'bar')
```

```python
print(len(X_test_50))
print(X_test_50.groupby('label').size().reset_index(name = 'count'))
print(X_test_50.head())

X_test_50['label'].value_counts().plot(kind = 'bar')
```
